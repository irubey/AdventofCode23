```python
import re
```

```python
with open('input-2.txt', 'r') as file:
    lines = [line.strip() for line in file]


print(lines)
```

```python
all_games_data = []

for game in lines:
    #Extract game number
    game_number_match = re.search(r'Game (\d+):', game)
    game_number = int(game_number_match.group(1))
    
    #Split the string into rounds
    rounds = game.split(';')

    #Process each round
    game_data = []
    for i, round in enumerate(rounds, start=1):
        #Extract number-color combinations
        number_color_pairs = re.findall(r'(\d+) (\w+)', round)

        #Append round number and number-color paors to the game data
        game_data.append({'game': game_number, 'round': i, 'pairs': number_color_pairs})
    
    all_games_data += game_data

print(all_games_data)
#game:#, round:#, pairs:[(#,color),]

```

```python
top_colors = {}

for game in set(d['game'] for d in all_games_data):  # Iterate through each unique game
    top_red = top_green = top_blue = 0

    for dictionary_item in filter(lambda d: d['game'] == game, all_games_data):
        pairs = dictionary_item['pairs']

        for number, color in pairs:
            number = int(number)
            if color == 'red' and number > top_red:
                top_red = number
            elif color == 'blue' and number > top_blue:
                top_blue = number
            elif color == 'green' and number > top_green:
                top_green = number

    top_colors[game] = (top_red, top_green, top_blue)

print(top_colors)
```

```python
products_dict = {
    game: red * green * blue for game, (red, green, blue) in top_colors.items()
}

total = sum(products_dict.values())
print(total)
```

```python
red_cube_real = 12
green_cube_real = 13
blue_cube_real = 14

game_failed = {}

for dictionary_item in all_games_data:
    game = dictionary_item['game']
    game_round = dictionary_item['round']
    pairs = dictionary_item['pairs']

    for number, color in pairs:
        if (color == 'red' and int(number) > red_cube_real) or \
            (color =='blue' and int(number) > blue_cube_real) or \
            (color == 'green' and int(number) > green_cube_real):
            game_failed[game] = game_round
            break


print(game_failed)


```

```python
passed_games = [game_dict for game_dict in all_games_data if game_dict.get('game') not in game_failed]


unique_game_ids = {game_dict['game'] for game_dict in passed_games}
ids = list(unique_game_ids)

print(sum(ids))

```
